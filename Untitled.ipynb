```python
# Import necessary libraries
import pandas as pd  # For data manipulation and analysis
import numpy as np  # For numerical operations
# Adjust pandas settings to display all columns
pd.set_option('display.max_columns', None)  # Show all columns
```

```python
# Step 1: Load the datasets
# Load the two Airbnb datasets into dataframes
airbnb1 = pd.read_csv("airbnb_1.csv")  
airbnb2 = pd.read_csv("airbnb_2.csv")
```

```python
# Display the column names of each dataset
print("\nColumn names in airbnb1:\n\n", airbnb1.columns.tolist())
print("\nColumn names in airbnb2:\n\n", airbnb2.columns.tolist())
```

```python

# Step 3: Merge the datasets
# Merge the two datasets. Assuming they share common columns; modify join parameters if needed.
merged_data = pd.concat([airbnb1, airbnb2], ignore_index=True)
# Display the shape of the merged dataset
print("\nShape of the merged dataset:", merged_data.shape)
```

```python
# Columns selected based on minimal missing data and relevance
columns_to_keep = [
    'id', 'name', 'description', 'host_id', 'host_name', 'host_since',
    'neighbourhood', 'neighbourhood_cleansed', 'latitude', 'longitude',
    'property_type', 'room_type', 'accommodates', 'bathrooms_text', 'bedrooms', 'beds',
    'amenities', 'price', 'minimum_nights', 'maximum_nights',
    'availability_365', 'number_of_reviews', 'review_scores_rating', 'reviews_per_month',
    'instant_bookable', 'calculated_host_listings_count', 'calculated_host_listings_count_entire_homes',
    'review_scores_cleanliness', 'review_scores_location'
]
```

```python
# Retain only the selected columns
filtered_data = merged_data[columns_to_keep]

# Display the shape of the dataset after column selection
print("Dataset shape after column selection:", filtered_data.shape)
```

```python
# Display the first few rows of the filtered dataset
print("\nFiltered dataset preview:")
filtered_data.head()
```

```python
# Check for missing values in the merged dataset
print("\nMissing values in each column:")
print(filtered_data.isnull().sum())
```

```python
# Handle missing values
# Drop columns with excessive missing data (if still any)
threshold = 0.3 * len(filtered_data)  # 30% missing value threshold
filtered_data = filtered_data.dropna(axis=1, thresh=threshold)

# Fill missing values
for col in filtered_data.columns:
    if filtered_data[col].dtype == 'object':  # For categorical/text columns
        filtered_data[col].fillna(filtered_data[col].mode()[0], inplace=True)
    else:  # For numerical columns
        filtered_data[col].fillna(filtered_data[col].median(), inplace=True)
```

```python
# Check for missing values after handling
print("\nMissing values after handling:")
print(filtered_data.isnull().sum())
```

```python
# Check and remove duplicate rows
print("\nNumber of duplicate rows:", filtered_data.duplicated().sum())
filtered_data = filtered_data.drop_duplicates()
```

```python
# Convert columns to appropriate data types
# Convert 'price' to numeric
filtered_data['price'] = filtered_data['price'].replace('[\$,]', '', regex=True).astype(float)
```

```python
# Convert 'host_since' to datetime
filtered_data['host_since'] = pd.to_datetime(filtered_data['host_since'])
```

```python
# Standardize text columns (e.g., amenities)
# Create a new column to indicate if "WiFi" is listed in amenities
filtered_data['has_wifi'] = filtered_data['amenities'].apply(lambda x: 1 if 'WiFi' in x else 0)
```

```python
# Display the final shape of the cleaned dataset
print("\nFinal dataset shape:", filtered_data.shape)
```

```python
# Preview the cleaned dataset
print("\nPreview of cleaned dataset:")
filtered_data.head()
```

```python
# Step 3: Data Exploration

import matplotlib.pyplot as plt
import seaborn as sns

# 1. Generate summary statistics for numeric columns
numeric_columns = [
    'latitude', 'longitude', 'accommodates', 'bedrooms', 'beds', 'price',
    'minimum_nights', 'maximum_nights', 'availability_365', 'number_of_reviews',
    'review_scores_rating', 'reviews_per_month', 'calculated_host_listings_count',
    'calculated_host_listings_count_entire_homes', 'review_scores_cleanliness',
    'review_scores_location'
]
```

```python
# Summary statistics
print("Summary statistics for numeric columns:\n\n")
filtered_data[numeric_columns].describe()
```

```python
# 2. Plot the distribution of key variables
key_columns = ['price', 'review_scores_rating', 'availability_365', 'number_of_reviews']

for column in key_columns:
    plt.figure(figsize=(8, 5))
    sns.histplot(filtered_data[column], kde=True, bins=30)
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```

```python
# 3. Explore relationships between variables
# Scatterplot: Price vs. Review Scores Rating
plt.figure(figsize=(8, 5))
sns.scatterplot(data=filtered_data, x='review_scores_rating', y='price', alpha=0.6)
plt.title('Price vs. Review Scores Rating')
plt.xlabel('Review Scores Rating')
plt.ylabel('Price')
plt.show()
```

```python
# Boxplot: Price by Room Type
plt.figure(figsize=(8, 5))
sns.boxplot(data=filtered_data, x='room_type', y='price')
plt.title('Price by Room Type')
plt.xlabel('Room Type')
plt.ylabel('Price')
plt.show()
```

```python
# Correlation heatmap
correlation_matrix = filtered_data[numeric_columns].corr()

plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap of Numeric Variables')
plt.show()
```
