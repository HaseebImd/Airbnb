```python
# Import necessary libraries
import pandas as pd  # For data manipulation and analysis
import numpy as np  # For numerical operations
# Adjust pandas settings to display all columns
pd.set_option('display.max_columns', None)  # Show all columns
```

```python
# Step 1: Load the datasets
# Load the two Airbnb datasets into dataframes
airbnb1 = pd.read_csv("airbnb_1.csv")  
airbnb2 = pd.read_csv("airbnb_2.csv")
```

```python
# Step 2: Basic information about the datasets
# Display the shape (number of rows and columns) of each dataset
print("\nShape of airbnb1:", airbnb1.shape)
print("Shape of airbnb2:", airbnb2.shape)
```

```python
# Display the column names of each dataset
print("\nColumn names in airbnb1:\n\n", airbnb1.columns.tolist())
print("\nColumn names in airbnb2:\n\n", airbnb2.columns.tolist())
```

```python

# Step 3: Merge the datasets
# Merge the two datasets. Assuming they share common columns; modify join parameters if needed.
merged_data = pd.concat([airbnb1, airbnb2], ignore_index=True)
# Display the shape of the merged dataset
print("\nShape of the merged dataset:", merged_data.shape)
```
